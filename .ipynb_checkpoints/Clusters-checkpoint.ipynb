```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

from google.colab import files
```

```python
#Import the data
dataset = pd.read_csv("matches_bigger_int.csv", sep = ',')
print(dataset.columns.values)
```

```python
#Clustered into two groups using KMeans. KMeans is generally considered inappropriatte for categorical data (Henry 2016).
#However, it does not hurt to use it before moving to other methods, such as LCA.

features = ["%d" % i for i in range(0,10)]

kmeans = KMeans(n_clusters=2)
kmeans.fit(dataset[features])
print(kmeans)

labels = kmeans.predict(dataset[features])
```

```python
#Reduced to three dimensions for visualization
colmap = ['r', 'b', 'g', 'o']
colors = [colmap[i] for i in labels]

pca = PCA(n_components=3)
pca_result = pca.fit_transform(dataset[features].values)

dataset['pca-one'] = pca_result[:,0]
dataset['pca-two'] = pca_result[:,1] 
dataset['pca-three'] = pca_result[:,2]

print('Explained variation per principal component: {}'.format(pca.explained_variance_ratio_))
```

```python
#Two graphs viewing the data from two angles. As expected the data was devided into the losing and wining teams. However,
#increasing the number of clusters did not reveal any meaningful subgroups.

fig = plt.figure(figsize=plt.figaspect(0.5))

x = dataset['pca-one']
y = dataset['pca-two']
z = dataset['pca-three']

ax = fig.add_subplot(1,2,1, projection = '3d')
ax.scatter(x,y,z,c = colors)
plt.xlabel('X')
plt.ylabel('Y')


ay = fig.add_subplot(1,2,2, projection = '3d')
ay.scatter(y,x,z,c = colors)
plt.xlabel('Y')
plt.ylabel('X')

plt.show()
```

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(dataset[features].values)

dataset['pca-one-2'] = pca_result[:,0]
dataset['pca-two-2'] = pca_result[:,1] 
print('Explained variation per principal component: {}'.format(pca.explained_variance_ratio_))
```

```python
colmap = ['r', 'b', 'g', 'o']
colors = [colmap[i] for i in labels]

a = dataset['pca-one-2']
b = dataset['pca-two-2']

fig = plt.figure()
plt.scatter(a,b,c= colors)

plt.show()
```

```python
#Run in colab to download the final dataset
dataset.to_csv('reduced_data.csv')
files.download('reduced_data.csv')
```
